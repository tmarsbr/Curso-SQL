```python
# Calcular a média das idades não nulas
Cursor.execute(
    '''
    SELECT AVG(Idade) 
    FROM minha_tabela2 
    WHERE Idade IS NOT NULL
    '''
)

media_idade = Cursor.fetchone()[0]  # Obtém a média como um float

if media_idade is not None:
    # Arredonda a média para um valor inteiro (opcional, dependendo do seu caso de uso)
    media_idade = round(media_idade)

    # Atualizar os valores nulos com a média calculada
    Cursor.execute(
        '''
        UPDATE minha_tabela2
        SET Idade = ?
        WHERE Idade IS NULL
        ''', (media_idade,)
    )

    # Efetuar commit das mudanças se estiver usando uma conexão de banco de dados que requer isso
    conn.commit()

# Selecionar e imprimir todos os registros para verificar as mudanças
consulta = Cursor.execute('SELECT * FROM minha_tabela2').fetchall()
for linha in consulta:
    print(linha)
```

# count, avg, sum

```python
# a COUNT() função retorna o número de linhas que correspondem a um critério específico.
consulta = Cursor.execute(
    ''' 
    SELECT COUNT(*)
    FROM minha_tabela2
    '''
).fetchall()

print(consulta)
```

```python
# inner join
consulta = Cursor.execute(
    """
    SELECT *
    FROM tab_valor_vendas
    INNER JOIN tab_cadastro_vendedor1
    ON tab_valor_vendas.id = tab_cadastro_vendedor1.ID
    """
).fetchall()

for linha in consulta:
    print(linha)
```
